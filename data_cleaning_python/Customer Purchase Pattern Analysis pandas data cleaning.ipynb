```python
# Imputing missing values in Review Rating column with the median rating of the product category 
```

```python
df['Review Rating'] = df.groupby('Category')['Review Rating'].transform(lambda x: x.fillna(x.median()))
```

```python
# after replacing missing values with median values accordingly again verify is there any null value present in data
```

```python
df.isnull().sum()
```

```python
# Renaming columns accordingly for better readability and documentation
```

```python
df.columns = df.columns.str.lower()
df.columns = df.columns.str.replace(' ','_')
```

```python
# after renaming column verify column data
```

```python
df.columns
```

```python
# rename column name 'purchase_amount_(usd)' to 'purchase_amount' for better readability
```

```python
df.rename( columns={'purchase_amount_(usd)': 'purchase_amount'}, inplace=True)
```

```python
# create a new column age_group according age range
```

```python
labels = ['Young Adult', 'Adult', 'Middle-aged', 'Senior']
df['age_group'] = pd.qcut(df['age'], q=4, labels = labels)
```

```python
# viewing first 10 rows data after grouping
```

```python
df[['age','age_group']].head(10)
```

```python
# create new column purchase_frequency_days using map command
```

```python
frequency_mapping = {
    'Fortnightly': 14,
    'Weekly': 7,
    'Monthly': 30,
    'Quarterly': 120,
    'Bi-Weekly': 14,
    'Annually': 365,
    'Every 3 Months': 90
}

df['purchase_frequency_days'] = df['frequency_of_purchases'].map(frequency_mapping)
```

```python
# viewing first 10 rows data after mapping
```

```python
df[['purchase_frequency_days','frequency_of_purchases']].head(10)
```

```python
(df['discount_applied'] == df['promo_code_used']).all()
```

```python
# after found both column data was same then one column will be deleted using drop command
```

```python
df = df.drop('promo_code_used', axis=1)
```

```python
from sqlalchemy import create_engine

# Step 1: Connect to PostgreSQL
# Replace placeholders with your actual details
username = "postgres"      # default user
password = "root" # the password you set during installation
host = "localhost"         # if running locally
port = "5432"              # default PostgreSQL port
database = "customer_behaviour"    # the database you created in pgAdmin

engine = create_engine(f"postgresql+psycopg2://{username}:{password}@{host}:{port}/{database}")

# Step 2: Load DataFrame into PostgreSQL
table_name = "customer"   # choose any table name
df.to_sql(table_name, engine, if_exists="replace", index=False)

print(f"Data successfully loaded into table '{table_name}' in database '{database}'.")
```
